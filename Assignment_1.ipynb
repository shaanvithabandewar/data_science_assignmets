```python
# standardization
def standardize_column(column):
  mean = column.mean()
  std_dev = column.std()
  standardize_column = (column - mean) / std_dev
  return standardize_column
```

```python
# applying standardization to each column
for column in numerical_columns:
  df[column] = standardize_column(df[column])
  print(f"standardized {column}")
```

```python
# standardization
def standardize_column(column):
  mean = column.mean()
  std_dev = column.std()
  standardize_column = (column - mean) / std_dev
  return standardize_column
# applying standardization to each column
for column in numerical_columns:
  df[column] = standardize_column(df[column])
  print(df[numerical_columns].head()) # No changes required here
```

```python
# categorical value to dummy value

df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
df
```

```python

# Creating a binary column 'High Sales' based on 'Total Sales Value'
threshold = df['Total Sales Value'].mean()
df['High Sales'] = (df['Total Sales Value'] > threshold).astype(int)

print(df.head())
```
