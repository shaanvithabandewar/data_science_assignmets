```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler


anime_df = pd.read_csv('anime.csv')

# Data Preprocessing

# Fill missing 'rating' values with the mean rating
anime_df['rating'].fillna(anime_df['rating'].mean(), inplace=True)

# Fill missing 'genre' and 'type' values with 'Unknown'
anime_df['genre'].fillna('Unknown', inplace=True)
anime_df['type'].fillna('Unknown', inplace=True)

# Feature Extraction
anime_df['genre'] = anime_df['genre'].str.split(', ')
anime_genre_encoded = anime_df['genre'].str.join('|').str.get_dummies()

# Normalize the 'rating' feature
scaler = StandardScaler()
anime_df['rating_normalized'] = scaler.fit_transform(anime_df[['rating']])

# Combine the one-hot encoded genres and normalized ratings into a single DataFrame
anime_features = pd.concat([anime_genre_encoded, anime_df['rating_normalized']], axis=1)

# Display the first few rows of the transformed dataset
anime_features.head()
```
