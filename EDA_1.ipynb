```python
#data cleaning
import pandas as pd

# Load the dataset
df = pd.read_csv('Cardiotocographic.csv')

# Display the first few rows of the dataset
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Handle missing values (e.g., imputation or deletion)
df = df.dropna()  # or you can use df.fillna(value) for imputation

# Check data types
print(df.dtypes)

# Convert data types if necessary (e.g., numerical values stored as strings)
# df['column_name'] = df['column_name'].astype(float) # Example conversion

# Detect and treat outliers if necessary (simple example using IQR)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

# Final clean dataset overview
print(df.describe())
```

```python
# Statistical summary
summary = df.describe()

# Display summary
print(summary)

# Additional statistical insights (e.g., skewness, kurtosis)
skewness = df.skew()
kurtosis = df.kurtosis()

print("Skewness:\n", skewness)
print("Kurtosis:\n", kurtosis)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical variables
df.hist(figsize=(15, 10))
plt.show()

# Boxplots to visualize distributions and outliers
plt.figure(figsize=(15, 10))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()

# Pair plots for relationships between variables
sns.pairplot(df)
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df.corr()

# Strong correlations (absolute value > 0.7)
strong_corr = correlation_matrix[abs(correlation_matrix) > 0.7]
print("Strong Correlations:\n", strong_corr)
```
